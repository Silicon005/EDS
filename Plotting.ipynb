```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = [1, 2, 3, 4, 5]  # x-axis values
y = [10, 15, 7, 12, 9]  # y-axis values

# Calculate density
density = np.random.randint(10, 100, len(x))

# Create a scatter plot with density
plt.scatter(x, y, s=density, alpha=0.7, cmap='viridis')

# Customize the plot
plt.title("Scatter Plot with Density")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a colorbar
cbar = plt.colorbar()
cbar.set_label('Density')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
sizes = [30, 25, 20, 15]
colors = ['red', 'orange', 'yellow', 'green']
explode = [0.1, 0, 0, 0]  # Explode the first slice

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Customize the chart
plt.title("Pie Chart Example")

# Display the chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
sizes = [30, 25, 20, 15]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
explode = [0.1, 0, 0, 0]  # Explode the first slice

# Create a pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors,
                                  autopct='%1.1f%%', startangle=90)

# Customize the chart
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
ax.set_title("Pie Chart Example")

# Add a legend with custom formatting
legend = ax.legend(wedges, labels, title='Fruits', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(legend.get_title(), fontsize='12')

# Add annotations inside the wedges
for i, autotext in enumerate(autotexts):
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_path_effects([plt.patheffects.Stroke(linewidth=1.5, foreground='black'),plt.patheffects.Normal()])

# Add a shadow to the wedges
for wedge in wedges:
    wedge.set_edgecolor('gray')
    wedge.set_linewidth(1)

# Display the chart
plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Group 1', 'Group 2', 'Group 3']
sizes_group1 = [30, 25, 20]
sizes_group2 = [15, 10, 5]
colors_group1 = ['#ff9999', '#ffc000', '#8fd9b6']
colors_group2 = ['#ff99ff', '#99ccff', '#ffcc99']
explode = [0.1, 0, 0]  # Explode the first slice in each group

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Create the first pie chart for Group 1
wedges_group1, _, _ = ax1.pie(sizes_group1, explode=explode, labels=labels, colors=colors_group1,
                              autopct='%1.1f%%', startangle=90)
ax1.set_title("Group 1")

# Create the second pie chart for Group 2
wedges_group2, _, _ = ax2.pie(sizes_group2, explode=explode, labels=labels, colors=colors_group2,
                              autopct='%1.1f%%', startangle=90)
ax2.set_title("Group 2")

# Customize the subplots
for ax, wedges in zip([ax1, ax2], [wedges_group1, wedges_group2]):
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
group1_values = [20, 25, 15]
group2_values = [10, 18, 22]

# Set the bar width
bar_width = 0.35

# Set the x positions of the bars
x = np.arange(len(categories))

# Create the figure and axis objects
fig, ax = plt.subplots()

# Create the bars for Group 1
ax.bar(x - bar_width/2, group1_values, bar_width, label='Group 1')

# Create the bars for Group 2
ax.bar(x + bar_width/2, group2_values, bar_width, label='Group 2')

# Customize the chart
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
group1_values = [20, 25, 15]
group2_values = [10, 18, 22]

# Set the bar width
bar_width = 0.35

# Set the x positions of the bars
x = range(len(categories))

# Create the grouped bar chart
plt.bar(x, group1_values, width=bar_width, label='Group 1')
plt.bar(x, group2_values, width=bar_width, label='Group 2', alpha=0.8)

# Customize the chart
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(x, categories)
plt.legend()

# Display the chart
plt.show()

```
