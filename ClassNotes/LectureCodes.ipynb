```python

f=open("empl1.csv","r")
content=f.read()
lines=content.split("\n")

eid=[]
name=[]
des=[]
sal=[]
for l in lines:
  words=l.split(",")
  print(words)
  eid.append(int(words[0]))
  name.append(words[1])
  des.append(words[2])
  sal.append(int(words[3]))


print("Employee Id",eid)
print("Employee Names",name)
print("Employee DEsignantion ",des)
print("Employee Salary",sal)

print("Maximum salary",max(sal))
print("Minimum salary",min(sal))
print("Sum of salary",sum(sal))
print("Average salary",sum(sal)/len(sal))

print("Employee with max salary", name[sal.index(max(sal))])
print("Employee with min salary", name[sal.index(min(sal))])


print("Sr managers are",end=" ")
for i in range(len(des)):
  if des[i]== "Sr. Manager":
    print(name[i])#remove end for new line



# prt employee name having salary 80000
print("employee with 80k",end=" ")
for i in range(len(sal)):
  if sal[i]== 80000:
    print(name[i])

#prt employee name salary 70000
print("employee with 100k",end=" ")
for i in range(len(sal)):
  if sal[i]== 100000:
    print(name[i],end=" ")
```
