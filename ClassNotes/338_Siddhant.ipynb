```python
import csv

def year_wise_employee_count():
    with open('data.csv', 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        employee_count = {}
        for row in reader:
            year = int(row[0])
            if year in employee_count:
                employee_count[year] += 1
            else:
                employee_count[year] = 1
        return employee_count

def average_salary_in_rupees():
    with open('data.csv', 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        total_salary = 0
        employee_count = 0
        for row in reader:
            salary = int(row[4])
            total_salary += salary
            employee_count += 1
        if employee_count > 0:
            average_salary = total_salary / employee_count
        else:
            average_salary = 0
        return average_salary

def medium_scale_company_count():
    with open('data.csv', 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        count = 0
        for row in reader:
            company_size = row[8]
            if company_size.lower() == 'm':
                count += 1
        return count

# Year-wise count of employees
employee_count_by_year = year_wise_employee_count()
print("Year-wise count of employees:")
for year, count in employee_count_by_year.items():
    print(f"Year {year}: {count} employees")

# Average salary in rupees
average_salary_rupees = average_salary_in_rupees()
print("Average salary in rupees:", average_salary_rupees)

# Number of employees working in a medium-scale company
medium_scale_company_count = medium_scale_company_count()
print("Number of employees working in a medium-scale company:", medium_scale_company_count)

```
