Copying and Viewing

Copying and Viewing


*  Copying : a new file will be created(copied)
*   Viewing : simply a shallow copy (viewing copy), the changes in the original file will be visible in the view file



```python
import numpy as np

# creating the array
arr = np.array([1,2,3])
print("Array =",arr)

# copy of the array
acopy = arr.copy()
print("Copy =",acopy)

# view of the array
aview = arr.view()
print("View =",aview)

# changing the array 
arr[0] = 69
print("Changed array =",arr)

# checking the changes
print("copy again =",acopy)
print("view again =",aview)
```

Stacking 1D arrays


*   vstack: stacks arrays in sequence vertically (row wise)
*   hstack: 
*   column_stack
*   row_stack





```python
import numpy as np

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

arr_vstack = np.vstack((arr1,arr2))
print("vstacked array =\n",arr_vstack)

arr_hstack = np.hstack((arr1,arr2))
print("hstacked array =\n",arr_hstack)

arr_rowstack = np.row_stack((arr1,arr2))
print("rowstacked array =\n",arr_rowstack)

arr_columnstack = np.column_stack((arr1,arr2))
print("columnstacked array =\n",arr_columnstack)
```

Data Sorting

```python
import numpy as np

arr1 = np.array([56,21,5,22])
arr2 = np.array([[2,3,1],[21,14,5]])
# sorted 1D array
print("sorted array =",np.sort(arr1))
# sorted 2D array
print("sorted array =\n",np.sort(arr2))

# reverse sorted 1Darray
print("sorted array in descending order =", np.sort(arr1)[::-1])
# reverse sorted 2D array
print("sorted array in descending order =\n", np.sort(arr2, axis=None, kind='quicksort')[::-1])
```

```python
# find the indexes where elements are common
a = np.array([1,2,3,4,5])
b = np.array([1,3,2,4,5])
print(np.where(a==b))


# finding the indexes of the even elements
arr = np.array([10,20,33,45,58,68,74,89])
x = np.where(arr%2 == 0)
print(x)
```

**searchsorted()** function : performs a binary search, and return the index of the given value from the given array


*   **values should be sorted first to get a correct output**



```python
import numpy as np

# searches for 7 in the sorted array
arr = np.array([6,7,8,9])
x = np.searchsorted(arr,7)
print(x)

# shows wrong inpute when values are not sorted
arr2 = np.array([34,89,12,9])
x = np.searchsorted(arr2,12)
print(x)
```

we can use the **side** parameter of **searchsorted()** function to specify the side from which we want the index

```python
import numpy as np

arr = np.array([6,7,8,9])
x = np.searchsorted(arr,7,side='right')
print(x)
```

searching for multiple values

```python
import numpy as np

arr = np.array([6,7,8,9])
x = np.searchsorted(arr,[7,8])
print(x)
```

**Broadcasting**: describes how NumPy treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes.
*   https://numpy.org/doc/1.20/user/theory.broadcasting.html
*   https://www.tutorialspoint.com/numpy/numpy_broadcasting.htm


```python
import numpy as np 
a = np.array([[0.0,0.0,0.0],[10.0,10.0,10.0],[20.0,20.0,20.0],[30.0,30.0,30.0]]) 
b = np.array([1.0,2.0,3.0])  
   
print("First array\n",a,"\n")
print("Second array\n",b,"\n")
print('First Array + Second Array\n',a+b,)
```

# **NUMPY STATISTICAL OPERATIONS**

Common Statistical Operations

```python
arr = np.array([1,2,3,4,5,9,6,7,8,9,9])

# standard deviation
print(np.std(arr))

# minimum
print(np.min(arr))

# summation
print(np.sum(arr))

# mean
print(np.mean(arr))

# variance
print(np.var(arr))

# mode (required scipy module)
import scipy
from scipy import stats
print("Most frequent element = ",stats.mode(arr)[0],"occurring",stats.mode(arr)[1],"times")
```

Here, in the above example, the order of 1st array is 4x3 and order of 2nd is 1x3. So the program broadcasts the 2nd array to become 4x3 by repeating the elements, and then the operation is done

Here, in the above example,  the orders of the arrays are different. Hence, the smaller array is broadcasted to be multiplied with the 1st array

**PRACTICAL3 **
1.   get a dataset (csv)
2.   get it into a list
3.   put that into an array
4.   and then perform the 5 numpy operations


```python
import numpy as np

# loading the csv file into an array using np.loadtext
arr = np.loadtxt("emp.csv",delimiter=',',dtype=str,skiprows=1)  
# skiprows parameter will skip the 1st row (header row)
print(arr,"\n")

#extracting data into lists
sal = []
exp = []
for i in arr:
    sal.append(int(i[2]))
    exp.append(int(i[3]))
print("Sal list:",sal)
print("Experience list:",exp)
print()

# converting lists to arrays
arr_sal = np.array(sal)
arr_exp = np.array(exp)
print("Sal Array: ",arr_sal)
print("Experience Array:",arr_exp)
print()


# NUMPY OPERATIONS
# 1) element-wise addition
print("1)Added array = ",arr_sal + arr_exp)

# 2) element-wise multiplications
print("2)Multiplied array =",arr_sal * arr_exp)

# 3) concatenation
combo = np.concatenate((arr_sal,arr_exp))
print("3)Combined array: ",combo)

# 4) reshaping
reshaped = np.reshape(combo,(4,4))
print("4)Reshaped array: ",reshaped)

# 5) mean
print("5)Mean Salary:",np.mean(arr_sal))

# 6) minimum experience
print("6)Min Exp:",np.min(arr_exp))
```
