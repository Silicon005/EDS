```python
import numpy as np
arr=np.array([10,20,5,7,6,8,7,75,54.7])
x=np.searchsorted(arr,7)
print(x)
```

```python
import numpy as np
arr=np.array([10,20,5,7,6,8,75,54.7])
x=np.searchsorted(arr,7,side='right')
print(x)
```

```python
array1=np.array([1,2,3,4,5,9,6,7,8,9,9])
#standard deviation
print(np.std(array1))
#minimum
print(np.min(array1))
#summation
print(np.sum(array1))
#median
print(np.median(array1))
#mean
print(np.mean(array1))
#mode
from scipy import stats
print("Most frequent element=",stats.mode(array1)[0])
print("Number of occurrences =",stats.mode(array1)[1])
#variance
print(np.var(array1))
```
