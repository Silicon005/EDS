```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('coffee.csv')
print(data.shape)
data.head()
```

```python
#Get x & y values
X=data['Aroma'].values
Y=data['Flavor'].values
mean_X=np.mean (X)
mean_Y=np.mean (Y)
n=len(X)
num=0
den=0
for i in range (n):
   num+=(X[i]-mean_X) * (Y[i]-mean_Y)
  # summation of (x-x')*(y-y') i.e. numerator in slope formula
   den+=(X[i]-mean_X) ** 2
  # summation of (x-x')^2 i.e. denominator
m=num/den
c=mean_Y-(m* mean_X)
print(m)
print(c)
```

```python
max_x=np.max(X)+1
min_x=np.min(X)-1
x=np.linspace(min_x, max_x, 1000)
y=m*x+c

plt.plot(x,y,color='g', label='Regression Line')
plt.scatter(X,Y,color='b', label='Scatter plot')

plt.xlabel('Aroma')
plt.ylabel('Flavor')
plt.legend(loc='lower right')
plt.show()
```

#R^2 = [sum(Yp-ybar)^2]/[sum(y-ybar)^2]

```python
sum1=0
sum2=0
for i in range(n):
  yp=m*X[i]+ c
  sum1+=(Y[i]-yp) **2
  sum2+=(Y[i]-mean_Y)**2
r2=1-(sum1/sum2)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=X.reshape((n,1))
reg=LinearRegression()

reg=reg.fit(X,Y)

r2_score=reg.score(X,Y)
print(r2_score)
```
